Øvelse 1: skriv en funktion lige_ulige(tal1,tal2) som tager to tal og returnerer en liste af tupler med hvert tal i listen i range(tal1, tal2). Brug list comprehension i funktionen. Eks. resultat af funktionkald lige_ulige(2,6) skal returnere [(2, 'lige'), (3, 'ulige'), (4, 'lige'), (5, 'ulige')]. Bemærk at hvis tal1 er > tal 2 skal lige_ulige bruge tal2 som første tal i range og tal1 som første tal, dvs. ulige_tal(1,5) og ulige_lige(5,1) skal returnere samme resultat. 
tal % 2
 

```python
def lige_ulige(tal1,tal2):
    if tal1>tal2:
        tal1,tal2=tal2,tal1
    return [(x,'lige') if x%2==0 else (x,'ulige') for x in range(tal1,tal2)]
print(lige_ulige(1,5))
print(lige_ulige(5,1))
```

```python
def percentage(count, total):
    return 100 * count / total

print(percentage(text2.count('are'),len(text2)))
print(percentage(text2.count('think'),len(text2)))
```

unigrammer er de enkelte ord (tokens) i et korpus
bigrammer er par af ord i et tekst eller korpus.
Fx for teksten: "Jeg hedder Costanza og arbejder på KU." har jeg følgende bigrammer:
Jeg  hedder, hedder Costanza, Costanza  og, og arbejder, arbejder på, på KU, KU .,
Ligeledes har man trigrammer, 4-grammer osv. De kaldes n-grammer og anvendes til sprogmodellering.
I NLTK kan man danne bigrammer automatisk: bigrams(text)

```python
[print(el) for el in bigrams(text3[:10])]
```

```python
[print(x) for x in bigrams(text7[:10])]
```

```python
[print(x) for x in bigrams(text6[40:50])]
```
