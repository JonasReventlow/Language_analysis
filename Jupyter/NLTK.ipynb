Skriv en funktion lige_ulige(tal1,tal2) som tager to tal og returnerer en liste af tupler med hvert tial i listen i range(tal1,tal2)

```python
def lige_ulige(tal1,tal2):
    if tal1>tal2:
        tal1,tal2 = tal2,tal1
    return [(x,'lige') if x%2==0 else (x,'ulige') for x in range(tal1,tal2)]
    
```

```python
lige_ulige(1,5)
```

```python
lige_ulige(5,1)
```
